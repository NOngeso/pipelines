```r
library(ggplot2)
library(scales)
```

```r
synteny <- read.table("/data2/k821209/pascal/test.csv",sep='\t',header=T)
```

```r
head(synteny)
```

```r
ggplot(synteny,aes(x=ath_loc,y=X5,color=tha_chr)) + geom_point(size=1) + 
facet_grid(ath_chr ~ .) + ylim(0,10) + scale_x_continuous(labels = comma) +
theme(axis.text.x = element_text(angle = 60, hjust = 1)) + ylab('Ks value') + xlab('Genomic position') + ylim(0,3)
```

```r
png('ath.synteny.esa.ks.pos.plot.png', height=3000, width=5000, res = 300, units = 'px')
print(ggplot(synteny,aes(x=ath_loc,y=X5,color=tha_chr)) + geom_point(size=1) + 
facet_grid(ath_chr ~ .) + ylim(0,10) + scale_x_continuous(labels = comma) +
theme(axis.text.x = element_text(angle = 60, hjust = 1)) + ylab('Ks value') + xlab('Genomic position'))
dev.off()
```
