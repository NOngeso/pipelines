```python
from __future__ import print_function
import kang
```

```python
file_fa_ath = './PhytozomeV10_download/PhytozomeV10/Athaliana/annotation/Athaliana_167_TAIR10.cds_primaryTranscriptOnly.fa'
dicH2seq_ath = kang.Fasta2dic(file_fa_ath)
file_fa_esa = './PhytozomeV10_download/PhytozomeV10/Esalsugineum/annotation/Esalsugineum_173_v1.0.cds_primaryTranscriptOnly.fa'
dicH2seq_esa = kang.Fasta2dic(file_fa_esa)
file_fa_cre = './PhytozomeV10_download/PhytozomeV10/Creinhardtii/annotation/Creinhardtii_281_v5.5.cds_primaryTranscriptOnly.fa'
dicH2seq_cre = kang.Fasta2dic(file_fa_cre)
file_fa_aly = './PhytozomeV10_download/PhytozomeV10/Alyrata/annotation/Alyrata_107_v1.0.cds_primaryTranscriptOnly.fa'
dicH2seq_aly = kang.Fasta2dic(file_fa_aly)
```

```python
def get_SFIsite(dicH2seq):
    num_base_SFI = 13
    count = 0 
    for strHD in dicH2seq:
        seq = dicH2seq[strHD]
        intid = 0 
        for i in range(len(seq)):
            if i+num_base_SFI > len(seq):
                continue
            seq_part = seq[i:i+num_base_SFI]
            if seq_part[0:4].upper() == 'GGCC' and seq_part[-4:].upper() == 'GGCC':
                print (strHD, i,i+13,seq_part[0:4]+'-'+seq_part[4:9]+'-'+seq_part[9:], seq_part,intid,sep='\t')
                intid += 1
        if intid > 0:
            count += 1
    print('#','number of genes',count)
```

```python
get_SFIsite(dicH2seq_ath)
```

```python
get_SFIsite(dicH2seq_esa)
```
