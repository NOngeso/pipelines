```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import pickle as pk
matplotlib.style.use('ggplot') 
```

```python
df_gff_gene_index = pd.read_pickle('tophat/df.cre.gff.gene.index.pickle')
df_gff_gene       = df_gff_gene_index.reset_index()
```

```python
df_gff_ath = pd.read_csv('../DroughtNet/PhytozomeV10_download/PhytozomeV10/Athaliana/annotation/Athaliana_167_TAIR10.gene.gff3',sep='\t',skiprows=2,header=None)
```

```python
mask = (df_gff_ath[2] == 'gene')
df_gff_gene_ath = df_gff_ath[mask]
```

```python
df_gff_gene_ath = df_gff_gene_ath.reset_index()
```

```python
df_gff_gene_ath
```

```python
def get_intergenic_list(df):
    list_intergenic = []
    for i in df.index:
        try:
            chr_left  = df.loc[i][0] 
            chr_right = df.loc[i+1][0] 
            if chr_left != chr_right:
                continue
            end   = df.loc[i][4]
            start = df.loc[i+1][3]
            list_intergenic.append(start - end)
        except KeyError:
            pass
    return list_intergenic
```

```python
list_intergenic_cre = get_intergenic_list(df_gff_gene)
list_intergenic_ath = get_intergenic_list(df_gff_gene_ath)
```

```python
list_intergenic_ath
```

```python
def get_hist(ax1,array1,xlim,lab):
    a1,b1,c1 = ax1.hist(array1,bins=np.arange(0,max(array1),5),histtype='step',alpha=0.7,label=lab)
    ax1.set_xlim(0,xlim)
    ax1.set_title('Intergenic region comparison')
    ax1.set_xlabel('Length (bp)')
    ax1.set_ylabel('Count')
    
```

```python
fig, ax1 = plt.subplots(1)
get_hist(ax1,list_intergenic_ath,3000,'Ath')
get_hist(ax1,list_intergenic_cre,3000,'Cre')
plt.legend()
```
